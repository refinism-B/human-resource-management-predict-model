```python
import pandas as pd
import numpy as np
import sklearn
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, hamming_loss
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# 載入模型

model = joblib.load('20250612_RFM.pkl')
```

```python
# 輸入資料函式

def project_matter():
    while True:
        matter = input('專案性質是？（請填進場/直播/錄製）')
        global stream, pre_setting, record
        if matter == '進場':
            stream = 0
            pre_setting = 1
            record = 0
            break

        elif matter == '直播':
            stream = 1
            pre_setting = 0
            record = 0
            break

        elif matter == '錄製':
            stream = 0
            pre_setting = 0
            record = 1
            break

        else:
            print('輸入錯誤請重新輸入')
```

```python
# 輸入預測資料

def Manual():
    month = input('幾月？')
    day = input('幾號？')
    weekday = input('星期幾？')
    holiday = input('是假日嗎？')
    time_len = input('活動多長？')
    cam = input('機位數有幾機？')
    b_roll = input('有花絮嗎？')
    switch = input('有視訊切換或視訊工程嗎？')
    concall = input('有視訊連線嗎？')
    pa = input('有負責PA音控嗎？')
    split = input('會一大場分為多個分場嗎？')
    project_matter()

    importing = {
        '月' : month,
        '日' : day,
        '星期' : weekday,
        '是否假日' : holiday,
        '時長' : time_len,
        '機位數量' : cam,
        '花絮' : b_roll,
        '視訊切換' : switch,
        '視訊連線' : concall,
        'PA音控' : pa,
        '大場分小場' : split,
        '工作性質_直播' : stream,
        '工作性質_進場' : pre_setting,
        '工作性質_錄製' : record,
    }

    global new_data
    new_data = pd.DataFrame([importing])
    new_data
```

```python
# 輸入csv檔案

def read_file():
    importing = pd.read_csv('test.csv')
    global new_data
    new_data = importing.drop(columns = ['專案', '日期', ])
```

```python
# 進行預測結果

while True:
    Q = input('如何輸入資料？（手動填寫/檔案匯入）')
    if Q == '手動填寫':
        Manual()
        break
    elif Q == '檔案匯入':
        read_file()
        break
    else:
        print('輸入錯誤，請重新輸入')

pred_data = model.predict(new_data)
pred_col = ['導播人數', '攝影人數', '音控人數', '直播人數', '機動人數', '花絮人數', '視訊切換人數', '視訊連線人數',
       '人數']

output = pd.DataFrame(pred_data, columns = pred_col)
print('--------------------')
print('預測結果：')
output
```
